# Apply Rinse Detection and Generate Output

```python
naam = 0

for file in files:

    val    = file
    
    
    levels = val['level']
    raws   = val['raw']
    raw_diff = []
    
    i = 1

    while(i < len(raws)):
        diff = raws.iloc[i]-raws.iloc[i-1]
        raw_diff.insert(len(raw_diff), diff)
        i += 1

    algo_output = []
    i = 0

    high_val = 0
    high_raw = 0
    
    session_start = False
    session_start_delay = 0
    
    

    while(i < len(raw_diff)):
        diff = raw_diff[i]
        raw  = raws.iloc[i+1]

        out_val = 0   

#         if (high_val != 0):
#             out_val = -3    
        
        if ( (i == high_val+3) and (session_start) ) :
            print("file", naam) 
            print("raw", raw) 
            print("high_raw", high_raw) 
            print("15%", high_raw - high_raw*0.11)
            
            
            if (raw < (high_raw - high_raw*0.11)):
                out_val = 10
            else:
                out_val = 5

            high_val = 0
            high_raw = 0

    
        if (session_start):
            session_start_delay += 1
            
       #print(session_start_delay)
        
        #and (high_val == 0)
        if ( (diff > 88)  ):
            
            out_val = 1
            
            print(naam, "--- raw", raw, "i", i)
            if not(session_start):
                session_start = True
                out_val = 2
                
            elif( (session_start_delay > 10)):
                out_val = 3
                
                if (raw > high_raw):                                
                    high_val = i
                    high_raw = raw
                    out_val = 4

        algo_output.insert(len(algo_output), out_val)
        i += 1


    raw_fig = plt.figure(figsize=(10, 10))
    plt.plot(levels)
    plt.plot(raws)
    plt.plot(raw_diff)
    
    raw_fig.savefig(str(naam) + '_0raws.png')
    plt.close()

    algo_fig = plt.figure(figsize=(10, 10))
    plt.plot(algo_output)
    
    algo_fig.savefig(str(naam) + '_algo.png')
    
    plt.close()
    
    naam += 1
    

```
