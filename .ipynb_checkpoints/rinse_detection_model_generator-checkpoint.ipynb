# Create Model

```python
upper_bound = []
lower_bound = []

file_i = files[0]['col2 '] #use first sample as data size template

scaled_data = []

#generalise data for purpose of extracting shape 
for file in files:   
    temp = trim_rinse_data(file['col2 '])
    temp = scale_data(temp)
    scaled_data.append(temp)
    
#"""  
plt.figure()
for a in scaled_data:
    #print(len(a))
    plt.plot(a)#"""
    
upper_bound = scaled_data[0].copy()
lower_bound = scaled_data[0].copy()

outside_bounds = 0

for data in scaled_data:
    i=0
    while ((i<len(data)-1) or (i<len(upper_bound)-1)):

        if (data[i] > upper_bound[i]):
            upper_bound[i] = data[i]
            
        if (data[i] < lower_bound[i]):
            lower_bound[i] = data[i]
                
        i+=1
print(upper_bound)
print(lower_bound)


plt.figure()
plt.plot(upper_bound)
plt.plot(lower_bound)
```

# Export Model into CSV and  C array

```python
output_model = np.transpose([upper_bound, lower_bound])

save_dir = "rinse_model.csv"
np.savetxt(save_dir, output_model,fmt="%i" , delimiter=",")
print("Saved CSV to " + save_dir)

#format model for C 2d array int array

array_size = str(len(output_model[0]))
array_length =  str(len(output_model))


c_rinse_model  = "#define MODEL_HEIGHT       " +str(model_height) + "\n"
c_rinse_model += "#define RINSE_MODEL_UPPER  0 \n"
c_rinse_model += "#define RINSE_MODEL_LOWER  1 \n"
c_rinse_model += "#define RINSE_MODEL_SIZE   " + array_size + "\n"
c_rinse_model += "#define RINSE_MODEL_LENGTH " + array_length  + "\n\n"

c_rinse_model += "int rinse_model [RINSE_MODEL_LENGTH][RINSE_MODEL_SIZE] = {"

for vals in output_model:
    c_rinse_model += "{" + str(vals[0]) + "," + str(vals[1]) + "},"

c_rinse_model = c_rinse_model[0:len(c_rinse_model)-1]
c_rinse_model += "};\n"

c_save_dir = "c_rinse_model.txt"

text_file = open(c_save_dir, "w")
n = text_file.write(c_rinse_model)
text_file.close()

print("Saved C Array to " + c_save_dir)


print((c_rinse_model))

```
